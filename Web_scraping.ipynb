```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.physio-pedia.com/Lymphoedema"
```

```python
req=requests.get(url)
```

```python
req.content
```

```python
# req=requests.get("https://www.physio-pedia.com/Lymphoedema")
```

%%%%%%%%%%%%%%%%Content%%%%%%%%%


```python
soup=BeautifulSoup(req.content,"html.parser")
```

```python
soup
```

```python
print(soup.prettify())
```

```python
#to get the proper data. to get ans for our question.
print(soup.get_text())
```
